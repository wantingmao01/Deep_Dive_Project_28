```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def getfile(location_pair,**kwargs): #tries to get local version and then defaults to google drive version
    (loc,gdrive)=location_pair
    try:
        out=pd.read_csv(loc,**kwargs)
    except FileNotFoundError:
        print("local file not found; accessing Google Drive")
        loc = 'https://drive.google.com/uc?export=download&id='+gdrive.split('/')[-2]
        out=pd.read_csv(loc,**kwargs)
    return out
```

```python
fname=("uiuc-gpa-dataset.csv","https://drive.google.com/file/d/1-NkhQzXPhPfrLvhmGuyfuu9gHAiDDLpa/view?usp=sharing")
data=getfile(fname)
data.head()
```

```python
data.columns
```

Calculate average gpa of each course.

```python
data["Class Size"] = data['A+'] + data['A'] + data['A-'] + data['B+'] + data['B'] + data['B-'] + data['C+'] + data['C'] + data['C-'] + data['D+'] + data['D'] + data['D-'] + data['F'] # W is not included as it does not affect GPA
data["Average GPA"] = (data['A+']*4 + data['A']*4 + data['A-']*3.67 + data['B+']*3.33 + data['B']*3 + data['B-']*2.67 + data['C+']*2.33 + data['C']*2.00 + data['C-']*1.67 + data['D+']*1.33 + data['D'] + data['D-']*0.67) / data["Class Size"]
```
